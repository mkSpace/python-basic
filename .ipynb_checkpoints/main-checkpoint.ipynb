```python
import pickle

# data = {
#     'offices': [[0, "New Engineering Building", "4105", "Jaemin"]],
#     'areas': [[0, 0, "Desk#1", "Jaemin"]],
#     'products': [[0, 0, 0, "PC", "Jaemin's PC", 2150000, "Jaemin", "20220101", "20221201"]]
#     'officesIncrementalKey': 0,
#     'areasIncrementalKey': 0,
#     'productsIncrementalKey': 0
# }
data = {}


def initializeData():
    global data
    data = {
        'offices': [[0, "New Engineering Building", "4105", "Jaemin"], [1, "Old Engineering Building", "4001", "Jaemin"]],
        'areas': [[0, 0, "Desk#1", "Jaemin"], [1, 1, "Desk#1", "Jaemin"], [2, 1, "Desk#2", "Jaemin"]],
        'products': [],
        'officesIncrementalKey': 2,
        'areasIncrementalKey': 3,
        'productsIncrementalKey': 0
    }


def dumpData():
    global data
    with open('./list.dat', 'wb') as file:
        pickle.dump(data, file, pickle.HIGHEST_PROTOCOL)
    print("Successfully dumped data")
    

def loadData():
    global data
    try:
        with open('./list.dat', 'rb') as file:
            data = pickle.load(file)
        print("Successfully Loaded data")
    except FileNotFoundError:
        print("이전에 저장된 데이터가 없습니다.")


def printMainManual():
    print("*********************************")
    print("1. 사무실 관리 ")
    print("2. 영역 관리 ")
    print("3. 물품 관리 ")
    print("4. 검색 ")
    print("5. 불러오기 ")
    print("6. 저장하기 ")
    print("7. 종료 ")
    print("*********************************")


def retrieveOffices():
    global data
    if len(data['offices']) < 1:
        print("There is no offices")
        return
    print(data)
    print(data['offices'])


def getAreasByOffice(officeCode):
    areas = []
    for area in data['areas']:
        if area[1] == officeCode:
            areas.append(area)
    return areas


def retrieveAreasByOffice(officeCode):
    areas = getAreasByOffice(officeCode)
    print(areas)


def retrieveAreas():
    print(data['areas'])


def printProductManual():
    print("*물품 관리 made by 2017112622 박재민*")
    print("1. 물품 조회 ")
    print("2. 물품 입력 ")
    print("3. 물품 수정 ")
    print("4. 물품 삭제 ")
    print("5. 물품 대여 ")
    print("6. 물품 이관 ")
    print("7. 뒤로 가기 ")
    print("******************************")


def retrieveProducts():
    print("*********************************")
    print("ProductCode  OfficeCode  AreaCode  ProductType  ProductName     Price    Lender   RentalDate   ReturnDate")
    for product in data['products']:
        print('%d\t     %d\t\t %d    \t   %s\t         %s\t%d\t%s\t%s\t%s' % (
            product[0], product[1], product[2], product[3], product[4], product[5], product[6], product[7], product[8]))


def inputProduct():
    global data
    if len(data['offices']) < 1:
        print("There is no offices")
        return
    print("*********Input Product**********")
    retrieveOffices()
    officeCode = int(input("Office Code : "))
    if len(getAreasByOffice(officeCode)) < 1:
        print("There is no areas")
        return
    retrieveAreasByOffice(officeCode)
    areaCode = int(input("Area Code : "))
    productType = input("Product Type : ")
    productName = input("Product Name : ")
    price = int(input("Price : "))
    while price < 0:
        print("Price must be equal or up to 0")
        price = int(input("Price : "))
    data['products'].append(
        [data['productsIncrementalKey'], officeCode, areaCode, productType, productName, price, "", "", ""]
    )
    data['productsIncrementalKey'] += 1


def updateProduct():
    global data
    if len(data['products']) < 1:
        print("There is no products")
        return
    retrieveProducts()
    productCode = int(input("Product Code : "))
    prevProduct = []
    prevProductIndex = -1
    for index, product in enumerate(data['products']):
        if product[0] == productCode:
            prevProductIndex = index
            prevProduct = product
            break
    if len(prevProduct) < 1 or prevProductIndex == -1:
        print("Cannot find the product code : ", productCode)
        return
    productType = input("Product Type : ")
    productName = input("Product Name : ")
    price = int(input("Price : "))
    while price < 0:
        print("Price must be equal or up to 0")
        price = int(input("Price : "))
    prevProduct[3] = productType
    prevProduct[4] = productName
    prevProduct[5] = price
    data['products'][prevProductIndex] = prevProduct
    print("Updated")
    retrieveProducts()


def deleteProduct():
    global data
    if len(data['products']) < 1:
        print("There is no products")
        return
    retrieveProducts()
    productCode = int(input("Product Code : "))
    target = []
    for product in data['products']:
        if product[0] == productCode:
            target = product
            break
    if len(target) < 1:
        print("Cannot find product code : ", productCode)
        return
    data['products'].remove(target)
    print("Deleted.")
    retrieveProducts()


def bookProduct():
    global data
    if len(data['products']) < 1:
        print("There is no products")
        return
    retrieveProducts()
    productCode = int(input("Product Code : "))
    productIndex = -1
    target = []
    for index, product in enumerate(data['products']):
        if product[0] == productCode:
            productIndex = index
            target = product
            break
    if len(target) < 1 or productIndex == -1:
        print("Cannot find product code : ", productCode)
        return
    if target[6].__len__() == 0 or target[8].__len__() != 0:
        lender = input("Lender : ")
        rentalDate = input("Rental Date : ")
        target[6] = lender
        target[7] = rentalDate
        data['products'][productIndex] = target
        print("Successfully Rentalled!!")
        retrieveProducts()
    else:
        print("Already rentalled")
        return


def moveProduct():
    global data
    if len(data['products']) < 1:
        print("There is no products")
        return
    if len(data['offices']) < 1:
        print("There is no offices")
        return
    retrieveProducts()
    productCode = int(input("Product Code : "))
    productIndex = -1
    target = []
    for index, product in enumerate(data['products']):
        if product[0] == productCode:
            productIndex = index
            target = product
            break
    if len(target) < 1 or productIndex == -1:
        print("Cannot find product code : ", productCode)
        return
    retrieveOffices()
    officeCode = int(input("Office Code : "))
    if len(getAreasByOffice(officeCode)) < 1:
        print("There is no areas")
        return
    retrieveAreasByOffice(officeCode)
    areaCode = int(input("Area Code : "))
    target[1] = officeCode
    target[2] = areaCode
    data['products'][productIndex] = target
    print("Successfully moved")
    retrieveProducts()
    

def printSearchProductManual():
    print("*물품 검색 made by 2017112622 박재민*")
    print("1. 물품 전체 조회")
    print("2. 물품 검색")
    print("3. 뒤로 가기")
    print("******************************")
    
    
def searchAllProducts():
    global data
    for office in data['offices']:
        print("== Office#%d %s %s Managed by %s ==" % (office[0], office[1], office[2], office[3]))
        for area in data['areas']:
            if area[1] != office[0]:
                continue
            print("==== Area#%d %s Managed by %s ====" % (area[0], area[2], area[3]))
            products = []
            for product in data['products']:
                if product[1] == office[0] and product[2] == area[0]:
                    products.append(product)
            if len(products) < 1:
                print("There is no products.", end='\n\n')
            else :
                print("====== Product List ======")
                for product in products:
                    print("======== Product#%d %s ========" % (product[0], product[4]))
                print("")
        

def searchProductByCode():
    global data
    productCode = int(input("Product code > "))
    for product in data['products']:
        if product[0] == productCode:
            print("ProductCode  OfficeCode  AreaCode  ProductType  ProductName     Price    Lender   RentalDate   ReturnDate")
            print('%d\t     %d\t\t %d    \t   %s\t         %s\t%d\t%s\t%s\t%s' % (product[0], product[1], product[2], product[3], product[4], product[5], product[6], product[7], product[8]))
            break


if __name__ == '__main__':
    initializeData()
    while True:
        printMainManual()
        mode = int(input("> "))
        if mode == 1:
            pass
        elif mode == 2:
            pass
        elif mode == 3:
            while True:
                printProductManual()
                productMode = int(input("> "))
                if productMode == 1:
                    retrieveProducts()
                elif productMode == 2:
                    inputProduct()
                elif productMode == 3:
                    updateProduct()
                elif productMode == 4:
                    deleteProduct()
                elif productMode == 5:
                    bookProduct()
                elif productMode == 6:
                    moveProduct()
                elif productMode == 7:
                    break
        elif mode == 4:
            while True:
                printSearchProductManual()
                searchingProductMode = int(input("> "))
                if searchingProductMode == 1:
                    searchAllProducts()
                elif searchingProductMode == 2:
                    searchProductByCode()
                elif searchingProductMode == 3:
                    break
        elif mode == 5:
            loadData()
        elif mode == 6:
            dumpData()
        elif mode == 7:
            print("Bye!!")
            break
```
